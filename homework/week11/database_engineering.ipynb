```python
import csv
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from time import time
from datetime import datetime
import sqlite3
```

```python
cleanmeasurement_path = 'Resources/data/clean_measurement.csv'
clean_measurement = pd.read_csv(cleanmeasurement_path)
station_path = 'Resources/data/station.csv'
station_data = pd.read_csv(station_path)
```

```python
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String, Float, DateTime
```

```python
class Measurement(Base):
    __tablename__ = 'measurement_data'
    id = Column(Integer, primary_key = True)
    station = Column(String(255))
    date = Column(DateTime)
    precipitation = Column(Float)
    temperature = Column(Integer)
```

```python
class Station(Base):
    __tablename__ = 'station_data'
    id = Column(Integer, primary_key = True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
    location = Column(String(255))
```

```python
engine = create_engine('sqlite:///hawaii_sqlite.sqlite')
conn = engine.connect()
Base.metadata.create_all(conn)
```

```python
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
for i, row in clean_measurement.iterrows():
    print ()
    record = Measurement(**{
        'station' : row.station,
        'date' : datetime.strptime(row.date, '%Y-%m-%d').date(),
        'precipitation' : row.prcp,
        'temperature' : row.tobs
})
    session.add(record)
session.commit()
```

```python
for i, row in station_data.iterrows():
    print ()
    record = Station (**{
        'station' : row.station,
        'name' : row.name,
        'latitude' : row.latitude,
        'longitude' : row.longitude,
        'elevation' : row.elevation,
        'location': row.location
})
    session.add(record)
session.commit()
```
